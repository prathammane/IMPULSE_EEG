Import required libraries and start with the extraction of time domain parameters (mean, rms, energy etc.) we will get channel wise parameter and graph for each of the four classes

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Plot EEG signals for the data point
def plot_eeg_signals(sample):
    num_channels, num_points = sample.shape
    print(f"Plotting EEG Signals - Shape: {sample.shape}")

    # Individual channel plots
    plt.figure(figsize=(15, 10))
    for channel in range(num_channels):
        plt.subplot(5, 4, channel + 1)  # Create grid layout for plots
        plt.plot(sample[channel])
        plt.title(f"Channel {channel + 1}")
        plt.tight_layout()
    plt.suptitle("Individual Channel Plots", y=1.02)
    plt.show()

    # Combined plot with all channels
    plt.figure(figsize=(12, 6))
    for channel in range(num_channels):
        plt.plot(sample[channel], label=f'Channel {channel + 1}')
    plt.title("Combined Plot of All Channels")
    plt.xlabel("Time Points")
    plt.ylabel("Amplitude")
    plt.legend(loc="upper right", bbox_to_anchor=(1.15, 1.0), fontsize=8)
    plt.show()

# Compute time-domain metrics for the data point
def compute_metrics(sample):
    metrics = {
        "Channel": [],
        "Mean": [],
        "Zero Crossing Rate": [],
        "Range": [],
        "Energy": [],
        "RMS": [],
        "Variance": []
    }
    for channel_idx, channel_data in enumerate(sample):
        metrics["Channel"].append(f"Channel {channel_idx + 1}")
        metrics["Mean"].append(np.mean(channel_data))
        metrics["Zero Crossing Rate"].append(((channel_data[:-1] * channel_data[1:]) < 0).sum() / len(channel_data)) #number of times its crossed 0/500
        metrics["Range"].append(np.ptp(channel_data))
        metrics["Energy"].append(np.sum(channel_data**2))
        metrics["RMS"].append(np.sqrt(np.mean(channel_data**2)))
        metrics["Variance"].append(np.var(channel_data))

    # Convert to DataFrame for better readability
    metrics_df = pd.DataFrame(metrics)
    return metrics_df
```

load and Pass a sample from complex partial seizures to the above function

```python
# Load the uploaded .npy file
dir="/content/drive/MyDrive/EEG_Data/train_data/Complex_Partial_Seizures/"
file_paths = ['0.npy']
for path in file_paths:
    sample = np.load(dir+path)

    # Check the shape of the loaded data
    print(f"Data shape: {sample.shape}")  # Should be (19, 500)

    # Plot signals
    plot_eeg_signals(sample)

    # Compute and display metrics
    print("Computing metrics for the uploaded sample")
    metrics_df = compute_metrics(sample)
    display(metrics_df)
```
